# Processing data - Apply NLP functions, identifying outliers

```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
file_path = 'master_df.csv' 
actual_data = pd.read_csv(file_path)

actual_data.head()
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

def clean_text(text):
    # Converting the text to lower case
    text = text.lower()
    # Removing special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Word tokenizing the text
    words = word_tokenize(text)
    # Removing stop words and lemmatizing each word
    words = [lemmatizer.lemmatize(word) for word in words if word not in stopwords.words('english')]
    return ' '.join(words)

# Applying the cleaning function to the 'Content' column
actual_data['Cleaned_Content'] = actual_data['content'].fillna('').apply(clean_text)

actual_data[['content', 'Cleaned_Content']].head()
```

```python
# Encoding 'Country' column
label_encoder = LabelEncoder()
actual_data['Country_Encoded'] = label_encoder.fit_transform(actual_data['country'])

# NLP Preprocessing on 'Cleaned_Content' column using TF-IDF
# Choosing 500 features for TF-IDF as a balanced approach
tfidf_vectorizer = TfidfVectorizer(max_features=500)
content_tfidf = tfidf_vectorizer.fit_transform(actual_data['Cleaned_Content'].fillna(''))

# Converting TF-IDF matrix to a DataFrame
tfidf_df = pd.DataFrame(content_tfidf.toarray(), columns=[f"TFIDF_{i}" for i in range(1, 501)])

# Adding the encoded and TF-IDF features to the original dataframe
processed_data = pd.concat([actual_data, tfidf_df], axis=1)


processed_data.head()
```

```python
import matplotlib.pyplot as plt

# List of columns for which we want to create boxplots
numerical_columns = ['gdp_per_capita', 'gdp_growth', 'gdp']

# Creating boxplots and calculating IQR
for column in numerical_columns:
    plt.figure(figsize=(8, 5))
    plt.boxplot(actual_data[column].dropna())
    plt.title(f'Boxplot for {column}')
    plt.ylabel(column)
    plt.show()

    # Calculating IQR
    Q1 = actual_data[column].quantile(0.25)
    Q3 = actual_data[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identifying outliers
    outliers = actual_data[(actual_data[column] < lower_bound) | (actual_data[column] > upper_bound)]

    print(f"For column '{column}':")
    print(f"IQR: {IQR}")
    print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    print(f"Number of outliers: {len(outliers)}")
    print()
```

```python
# Save the processed data to a CSV file
processed_data.to_csv('processed_data.csv', index=False)
```

# Outlier Removal and Traning the data set

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
data = pd.read_csv('processed_data.csv')
data.shape
```

```python
# Select the numerical columns
numeric_columns = ['gdp_per_capita', 'gdp_growth', 'gdp']

# Function for quantile-based flooring and capping
def quantile_flooring_capping(df, columns):
    
    for col in columns:
        floor = df[col].quantile(0.10)
        cap = df[col].quantile(0.90)
        df[col] = np.where(df[col]<floor, floor, df[col])
        df[col] = np.where(df[col]>cap, cap, df[col])

    return df

# Apply quantile-based flooring and capping
data_quantile_capped = quantile_flooring_capping(data.copy(), numeric_columns)
```

```python
# Function to trim outliers based on the 1st and 99th percentiles
def trim_outliers(df, columns):
    for col in columns:
        lower_bound = df[col].quantile(0.10)
        upper_bound = df[col].quantile(0.90)
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Apply trimming 
data_trimmed = trim_outliers(data.copy(), numeric_columns)
```

```python
# Function to apply log transformation
def log_transform(df, columns):
    for col in columns:
        df[col] = df[col].apply(lambda x: np.log1p(x) if x > 0 else 0) 
    return df

# Apply log transformation 
data_log_transformed = log_transform(data.copy(), numeric_columns)
```

```python
# Checking the number of rows after each method
capped_shape = data_quantile_capped.shape
trimmed_shape = data_trimmed.shape
log_transformed_shape = data_log_transformed.shape

capped_shape, trimmed_shape, log_transformed_shape
```

```python
# Function to plot histogram of all the numerical data
def plot_numerical_histograms(df):
    
    # Loop through each numerical column and plot histogram
    for col in numeric_columns:
        plt.figure(figsize=(8, 5))
        sns.histplot(df[col], bins=20, kde=True, color='blue')
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.grid(False)
        plt.show()
```

#### Observation
The histograms with KDE (Kernel Density Estimate) overlay for each numerical column show the distribution of values within those columns. Peaks in the histograms indicate common ranges for each feature, while the KDE helps in visualizing the underlying probability distribution

```python
# Select the TFIF columns
tfidf_columns = [col for col in data.columns if col.startswith('TFIDF')]

# Standardizing the features in the log-transformed dataset
scaler = StandardScaler()
scaled_features_log = scaler.fit_transform(data_log_transformed[tfidf_columns])

# Applying PCA for dimensionality reduction to 10 components
pca_log = PCA(n_components=10)
pca_features_log = pca_log.fit_transform(scaled_features_log)

# K-means clustering on the log-transformed data
kmeans_log = KMeans(n_clusters=3, random_state=42)
kmeans_labels_log = kmeans_log.fit_predict(pca_features_log)

# Hierarchical Clustering on the log-transformed data
hierarchical_log = AgglomerativeClustering(n_clusters=3)
hierarchical_labels_log = hierarchical_log.fit_predict(pca_features_log)

# Adding clustering labels to the log-transformed dataframe
data_log_transformed['kmeans_cluster'] = kmeans_labels_log
data_log_transformed['hierarchical_cluster'] = hierarchical_labels_log

# Visualizing the clustering results using PCA first two principal components
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.scatter(pca_features_log[:, 0], pca_features_log[:, 1], c=kmeans_labels_log, cmap='viridis')
plt.title('K-means Clustering on Log-Transformed Data')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

plt.subplot(1, 2, 2)
plt.scatter(pca_features_log[:, 0], pca_features_log[:, 1], c=hierarchical_labels_log, cmap='viridis')
plt.title('Hierarchical Clustering on Log-Transformed Data')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

plt.tight_layout()
plt.show()
```

#### Observation
The PCA visualization of clustering indicates that both K-means and hierarchical clustering algorithms are capturing similar patterns in the data after the log transformation and PCA dimensionality reduction. Both methods reveal three clusters, suggesting that the data structure is consistent and well-separated in the PCA space.


```python
# Selecting target and features for supervised learning
X = data_trimmed[tfidf_columns] 
y = data_trimmed['gdp']  

# Split the data into training and test sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

### Linear Regression
# Initialize and train the model
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)

# Make predictions on the test set
y_pred_lr = linear_reg.predict(X_test)

# Evaluate the model
mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)
print(f"Linear Regression - Mean Squared Error: {round(mse_lr, 2)}, R2 Score: {round(r2_lr, 2)}")

### Random Forest Regression
# Initialize and train the model
rf_reg = RandomForestRegressor(n_estimators=100, random_state=42)
rf_reg.fit(X_train, y_train)

# Make predictions on the test set
y_pred_rf = rf_reg.predict(X_test)

# Evaluate the model
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
print(f"Random Forest Regression - Mean Squared Error: {round(mse_rf, 2)}, R2 Score: {round(r2_rf, 2)}")
```

```python
# Plot Actual vs Predicted for Linear Regression
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x=y_test, y=y_pred_lr, color='blue', alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linewidth=2)
plt.title('Linear Regression: Actual vs Predicted')
plt.xlabel('Actual GDP')
plt.ylabel('Predicted GDP')

# Plot Actual vs Predicted for Random Forest Regression
plt.subplot(1, 2, 2)
sns.scatterplot(x=y_test, y=y_pred_rf, color='green', alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linewidth=2)
plt.title('Random Forest Regression: Actual vs Predicted')
plt.xlabel('Actual GDP')
plt.ylabel('Predicted GDP')

plt.tight_layout()
plt.show()

# Plot Residuals for Linear Regression
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(y_test - y_pred_lr, kde=True, color='blue', bins=30)
plt.title('Linear Regression: Residuals')
plt.xlabel('Residuals')

# Residuals for Random Forest Regression
plt.subplot(1, 2, 2)
sns.histplot(y_test - y_pred_rf, kde=True, color='green', bins=30)
plt.title('Random Forest Regression: Residuals')
plt.xlabel('Residuals')

plt.tight_layout()
plt.show()
```

#### Observation
The supervised learning models demonstrated varying levels of performance in predicting GDP-related outcomes. Linear Regression showed a simpler relationship between features and the target, whereas Random Forest Regression effectively captured non-linear patterns, resulting in better prediction accuracy and lower residuals. The comparison of actual vs. predicted values indicated that the Random Forest model achieved more consistent and closer predictions to the actual values, as reflected by its tighter clustering along the ideal prediction line.

The analysis of residuals highlighted that the Random Forest model had a more normally distributed and centered error, suggesting a more robust fit to the data compared to Linear Regression. This indicates the importance of model selection based on data characteristics, where ensemble methods like Random Forest can excel in capturing complex relationships that simpler models might miss.
